```python
import torch

def inspect(path):
    obj = torch.load(path, map_location='cpu')
    print(f'\n===== {path} =====')
    print('Type:', type(obj))

    # 情况 1：整个 nn.Module
    if isinstance(obj, torch.nn.Module):
        print('=> 是一个 nn.Module，打印第一层权重形状：')
        for name, param in obj.named_parameters():
            print(f'  {name}: {param.shape}')
            break  # 只看第一层
        # 如果是 RNN/LSTM，也可以直接看
        if hasattr(obj, 'memory'):
            print('LSTM input_size:', obj.memory.input_size)

    # 情况 2：state_dict
    elif isinstance(obj, dict):
        print('=> 是一个 dict(state_dict)，打印前 5 个 tensor 形状：')
        for k, v in list(obj.items())[:5]:
            print(f'  {k}: {v.shape}')
        # 如果带 LSTM 权重，也能看出来
        for k in obj.keys():
            if 'weight_ih' in k:  # LSTM 第一门权重
                print('LSTM input_size 推断:', obj[k].shape[1])
                break
    else:
        print('=> 未识别类型，内容：', obj)
    return obj

policy_raw = inspect("/home/guma_xiayu/xiayu/code/ASAP/deploy/policy/skill_cooldown/model/policy_15dof_raw.pt")
policy_now = inspect("/home/guma_xiayu/xiayu/code/ASAP/deploy/policy/skill_cooldown/model/policy_15dof.pt")
```
